**What does T mean?**
*   T  represents a linear transformation from a two-dimensional vector space (R^2) to itself.
*   Here, the expression T(x)=D.x scales the vector x, with the first row being scaled by 3 and the second row being scaled by 6, as determined by the matrix D.

**What does basis vectors mean?**


*   Basis vectors are linearly independent i.e, they can't be expressed as linear combination of any other vectors in the space.
*   However, every vector within the space can be expressed as linear combination of basis vectors.

Not always, but usually in Euclidean spaces(R^n), basis vectors are chosen to be orthogonal.



**To plot e1 and e2 vectors on a coordinate system**

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the vectors
e1 = np.array([[1], [0]])
e2 = np.array([[0], [1]])
```

```python
# Plot the e1, e2
plt.quiver(0, 0, e1[0], e1[1],angles='xy',scale_units='xy' ,scale=1, color='b', label='e1')
plt.quiver(0, 0, e2[0], e2[1], angles='xy', scale_units='xy', scale=1, color='r', label='e2')

# Set plot limits
plt.xlim(-1, 1.5)
plt.ylim(-1, 1.5)

# Add grid and labels
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of e1 and e2 vectors')
plt.legend()

# Show plot
plt.show()
```

**Transformation of e1 and e2 using the Diagonal matrix, D**

```python
# Declare D
D=np.array([[3,0],[0,6]])

# Matrix multiplication
Te1=np.dot(D, e1)
print("T(e1):")
print(Te1)
Te2=np.dot(D, e2)
print("T(e2):")
print(Te2)
```

```python
# Plot T(e1) and T(e2)
plt.quiver(0, 0, Te1[0], Te1[1], angles='xy', scale_units='xy' ,scale=1, color='b', label='T(e1)')
plt.quiver(0, 0, Te2[0], Te2[1], angles='xy', scale_units='xy', scale=1, color='r', label='T(e2)')

# Set plot limits
plt.xlim(-1, 5)
plt.ylim(-1, 8)

# Add grid and labels
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of T(e1) and T(e2) vectors')
plt.legend()

# Show plot
plt.show()
```

*Observation:*

*On transformation, e1 vector is scaled by 3 along x-axis and e2 vector by 6 along y-axis.*

*Only the magnitude varies, and the direction remains the same.*

**Transformation of vector, x using the Diagonal matrix, D**

```python
# Declare x
x=np.array([[1],[1]])
print("x:")
print(x)

# Transformation of x
Tx=np.dot(D, x)
print("T(x):")
print(Tx)
```

**Plot of transformed x**

```python
# Plot x and T(x)

plt.quiver(0, 0, Tx[0], Tx[1],angles='xy',scale_units='xy' ,scale=1, color='b', label='T(x)')
plt.quiver(0, 0, x[0], x[1], angles='xy', scale_units='xy', scale=1, color='r', label='x')

# Set plot limits
plt.xlim(-1, 5)
plt.ylim(-1, 8)

# Add grid and labels
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of x and T(x) vectors')
plt.legend()

# Show plot
plt.show()
```

*Observation:*

*Transformation of vector x, scales x-axis and y-axis by factor 3 and 6 respectively bringing about change in both the direction and magnitude.*

**Diagonal matrix:**

Must satisfy:


*   Square matrix.
*   Elements at diagonal must be non-zero while all other elements must be zero.



```python
D2=np.dot(D, D)
print("D:\n", D)
print("D2:\n", D2)
```

*D^2 represents the square of the original diagonal matrix D.*

i.e, When power of diagonal matrix is raised, each diagonal element is raised to that power while the off-diagonal elements remain zero.
