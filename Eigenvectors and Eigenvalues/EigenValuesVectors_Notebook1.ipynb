**Verifying eigenvector for a 2x2 matrix**

Solution:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the matrix A
A = np.array([[1, 6], [5, 2]])
print("Matrix A:")
print(A)
```

```python
# Define the eigenvectors u and v
u = np.array([[6],
              [-5]])
v = np.array([[3],
             [-2]])
print("Eigen vector u:",u)
print("Eigen vector v:",v)
```

```python
# Calculate Au and Av
Au = np.dot(A, u)
print("Au=",Au)
Av = np.dot(A, v)
print("Av=",Av)
```

```python
#Calculate eigenvalues
eigenvalues, _ = np.linalg.eig(A)
print(eigenvalues)
```

```python
#Check if u is an eigenvector
eigen_u = False
for eigenvalue in eigenvalues:
    if np.allclose(Au, eigenvalue * u):
        eigen_u = True
        break
if eigen_u:
    print("u is an eigenvector of A.")
else:
    print("u is not an eigenvector of A.")
```

```python
#Check if v is an eigenvector
eigen_v = False
for eigenvalue in eigenvalues:
    if np.allclose(Av, eigenvalue * v):
        eigen_v = True
        break
if eigen_v:
    print("v is an eigenvector of A.")
else:
    print("v is not an eigenvector of A.")
```

```python
plt.figure(figsize=(8, 6))
plt.plot([0, u[0,0]], [0, u[1,0]], 'b-', label='u')
plt.plot([0, v[0,0]], [0, v[1,0]], 'g-', label='v')
plt.plot([0, Au[0,0]], [0, Au[1,0]], 'b--', label='Au')
plt.plot([0, Av[0,0]], [0, Av[1,0]], 'g--', label='Av')

# Set plot limits
plt.xlim(-7.5, 7.5)
plt.ylim(-10, 10)

# Add grid and labels
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Action of A on u and v')
plt.legend()

# Show plot
plt.show()
```

**Direction of the eigenvector doesn't change**.

So, it is evident here that u is an eigenvector since it aligns parallelly with Au, whereas v isn't, as it is not parallel to Av and deviates.
